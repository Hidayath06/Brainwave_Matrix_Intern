```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# Manually specify the filename (if already uploaded)
filename = "Flipkart_Sales_.csv"  # Replace with your actual filename

# Read the CSV file
df = pd.read_csv(filename)

# Display first few rows
print(df.head())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Sales', data=df, marker='o', color='b')
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
top_products = df.groupby('Product')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_products.index, y=top_products.values, palette='coolwarm')
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Sales', data=df, palette='viridis')
plt.title('Category-wise Sales Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Discount (%)', y='Sales', data=df, alpha=0.7)
plt.title('Effect of Discount on Sales')
plt.xlabel('Discount (%)')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Rating'], bins=10, kde=True, color='purple')
plt.title('Product Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```
